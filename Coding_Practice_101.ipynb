# 1. Check a number is a perfect sq or not?

````python
import math
def perfect_sq(x:int):
  sq_rt = int(math.sqrt(x))
  if x < 0:
    print(False)
  if sq_rt * sq_rt == x:
    print(True)
  else: print(False)
````

````python
perfect_sq(7)
````

````python
import math
def is_pf(num:int):
  sr = math.sqrt(num)
  if num == sr*sr:
    print(True)
  else: print(False)
````

````python
is_pf(5)
````

````python
def merge_sort(a):
  temp = []
  for i in a:
    if i>i+1:
      temp.append(i+1)
      i+=1
    else:
      temp.append(i)
  return temp
````

# Second Largest
Given an array of positive integers arr[], return the second largest element from the array. If the second largest element doesn't exist then return -1.

````python
def getSecondLargest(arr):
    arr = list(set(arr))     # remove duplicates so we don't mistakenly pick the same largest twice
    arr.sort()               # sort the list in ascending order
    if len(arr) < 2:         # if fewer than 2 unique numbers, second largest does not exist
        return -1            # return -1 as per requirement
    return arr[-2]           # return the second last element (second largest number)
````

````python
ls = [2,2,2,2]
ls1 = [2,3,4,5,9,55,55,33,33]

print(getSecondLargest(ls))
getSecondLargest(ls1)
````

# Take a number and give sum of these numbers
```
num = int(input())              # Take a 3-digit integer
digits = str(num)               # Convert to string, e.g. 482 -> "482"
digit_list = map(int, digits)   # Convert each character to int -> [4, 8, 2]
result = sum(digit_list)        # Add them up
print(result)                   # Output: 14
```


```
map(function, iterable)
	•	function → A function you want to apply (built-in like int, str, abs, or your own custom function).
	•	iterable → Something you can loop over (like a list, tuple, set, or string).
	•	Result → Returns a map object (an iterator), where each element is the result of applying the function to the iterable’s elements.

# Example 1 – Simple with int

nums = ["1", "2", "3"]
mapped = map(int, nums)
print(list(mapped))   # [1, 2, 3]

#Example 2 – With a custom function

def square(x):
    return x * x

nums = [1, 2, 3, 4]
result = map(square, nums)
print(list(result))   # [1, 4, 9, 16]


# Example 3 – Using lambda (anonymous function)

nums = [1, 2, 3, 4]
result = map(lambda x: x + 10, nums)
print(list(result))   # [11, 12, 13, 14]
```

````python
# Fun program: Sum of digits of a 3-digit number 🎉

num = int(input("Enter a 3-digit number: "))

# Extract digits
d1 = num // 100        # Hundreds place
d2 = (num // 10) % 10  # Tens place
d3 = num % 10          # Ones place

# Calculate sum
digit_sum = d1 + d2 + d3

print(f"The sum of digits of {num} is: {digit_sum} 😃")
````
